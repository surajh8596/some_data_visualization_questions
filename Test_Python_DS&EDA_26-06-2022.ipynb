**2.Write a NumPy program to compute sum of all elements, sum of each column and sum of each row of a given array.**

```python
arr1=np.array([[1,3],[5,7]])
arr2=np.array([[2,4],[6,8]])
print('array1=', arr1)
print('array2=', arr2)
```

```python
arr1+arr2
```

```python
# sum of all elements(1D array)
arr=np.array([2,4,6,8,10])
sum=0
for i in arr:
    sum=sum+i
print(sum)
```

```python
arr=np.array([2,4,6,8,10])
sum_of_array = np.sum(arr)
print(sum_of_array)
```

```python
# sum of all elements(2D array)
arr=np.array([[1,3,5,7,9],[2,4,6,8,10]])
sum=0
for i in arr:
    for j in i:
        sum=sum+j
print(sum)
```

# Finding Top 10 products using pie chart in terms of quantity sold?

```python
pd.pivot_table(amazon, index='description', values='quantity', aggfunc=sum).sort_values(by='quantity', ascending=False)[:10]
```

```python
pd.pivot_table(amazon, 
               index='description', values='quantity', aggfunc=sum).sort_values(by='quantity',
                                                                                ascending=False)[:10].plot(kind='pie',
                                                                                                           subplots=True,
                                                                                                           legend=False,
                                                                                                           figsize=(20,20),
                                                                                                           autopct='%1.2f%%',
                                                                                                           fontsize=20);
```
